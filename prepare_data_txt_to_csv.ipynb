```python
import pandas as pd
from tqdm import tqdm_notebook
```

```python
file = open('twitter-2013train-A.txt')
```

```python
lines = file.readlines()
print('Number of lines in txt file: {}'.format(len(lines)))
```

```python
labels = []
tweets = []

for line in tqdm_notebook(lines):
    line = line.split()
    labels.append(line[1])
    line = line[2:]
    tweet = ' '.join(line)
    tweets.append(tweet)
```

```python
df = pd.DataFrame(list(zip(tweets, labels)), columns =['tweets', 'labels'])
```

```python
df.to_csv('tweets.csv', index=None)
```
